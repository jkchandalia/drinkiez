```python
import requests
import requests

import datetime
import json
import pandas as pd
```

```python
url = "https://www.thecocktaildb.com/api/json/v2/9973533/filter.php?a=Alcoholic"
response = requests.get(url)
print(response.json())
apiKey = 9973533
```

```python
output=response.json()['drinks']
```

```python
csv_out = []
for i, drink in enumerate(output):
    drink_line = [drink['strDrink'], drink['idDrink']]
    csv_out.append(drink_line)
    
```

```python
len(csv_out)
```

```python
for i, item in enumerate(csv_out):
    parameters = {"i": item[1]}
    print(item[1])

    lookup_request_url = "https://www.thecocktaildb.com/api/json/v2/"+str(apiKey)+"/lookup.php"
    response = requests.get(lookup_request_url,params=parameters)
    output=response.json()['drinks'][0]
    #print(output)
    for i in range(1,12):
        if output['strIngredient'+str(i)] is not None:
            item.append(output['strIngredient'+str(i)])
    csv_out[i]=item
        
```

```python
csv_out
long=[]
for drink in csv_out:
    for i in range(2,len(drink)):
        long.append([drink[1], drink[i]])
```

```python
len(long)
df_long=pd.DataFrame(long)
df_long[2]=1
```

```python
df_long.columns=["drinkId", "Ingredient", "present"]
print(len(df_long))
#
df_long=df_long.drop_duplicates()
len(df_long)
```

```python
df_wide=df_long.pivot(index='drinkId', columns='Ingredient', values='present')
df_wide.fillna(0)
df_wide=df_wide.iloc[:,1:].fillna(0)
```

```python
df_wide=df_wide.astype(int)
sum(df_wide.iloc[:,90])
df_distance = df_wide
df_distance.to_csv("df_distance.csv")
```

```python
def find_distance(drink_id, df_distance):
    distances=[]
    matrix=df_distance.as_matrix()
    index=list(df_distance.index).index(drink_id)
    for i in range(0,len(df_distance)):
        distances.append([list(df_distance.index)[i],np.linalg.norm(matrix[index,:]-matrix[i,:])])
    df_distances = pandas.DataFrame(distances)
    df_distances.columns=["id","dist"]
    df_distances=df_distances.sort_values(["dist"])
    return df_distances
        
def findTop5(df_distances):
    return df_distances['id'][1:6]
```
